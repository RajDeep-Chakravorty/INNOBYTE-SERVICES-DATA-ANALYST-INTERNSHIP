## About the features
- **Index**: Unique identifier for each record. Integer.
- **Order ID**: Unique identifier for each order. String.
- **Date**: Date when the order was placed. Date format - yyyy-DD-MM.
- **Status**: Current status of the order. String.
- **Fulfilment**: Method by which the order is fulfilled. String.
- **Sales Channel**: Platform through which the order was made. String.
- **Ship-Service-Level**: Level of shipping service chosen for the order. String.
- **Category**: Product category of the ordered item. String.
- **Size**: Size of the ordered item, if applicable. String.
- **Courier Status**: Current status of the order with the courier. String.
- **Currency**: Currency in which the order was placed. String.
- **Amount**: Total amount of the order. Float.
- **Ship-City**: City to which the order is shipped. String.
- **Ship-State**: State to which the order is shipped. String.
- **Ship-Postal-Code**: Postal code to which the order is shipped. String.
- **Ship-Country**: Country to which the order is shipped. String.
- **B2B**: Indicates if the order is a Business-to-Business order. Boolean.
- **Fulfilled-By**: Entity responsible for fulfilling the order. String.
- **New**: Indicates if the order is new. Boolean.
- **PendingS**: Indicates if the order is pending shipment. Boolean.

## 1. Sales Overview: Understand the overall sales performance, trends, and patterns over time

```python
df3 = df1.copy()
```

```python
df3['Date'] = df3['Date'].apply(pd.to_datetime)
df3.set_index('Date', inplace = True)
```

```python
df3.groupby('Category').Amount.sum()
```

```python
df3.groupby('Category').get_group(name = 'Shirt')['Amount'].sum()
```

```python
fig, ax = plt.subplots(9, 1, figsize = (15, 50))
# sns.scatterplot(x = data1.index,y = 'Amount', hue = 'Category', data = data1)
cat_list = df1.Category.unique().tolist()
cat_list = iter(cat_list)
for i in range(9):
    a = next(cat_list)
    df4 = df3[df3['Category'] == a].copy()
    axs = sns.lineplot(x = df4.index,y = 'Amount', data = df4, ax = ax[i])
    axs.set_title(f'Sales of {a}')
    axs.grid()
```

- As you can see in this analysis, the average sales of each category of product is best in between 01-05-2022 to 01-06-2022
- The watch has no sales as it is the least bought product in this span of time

## 2. Product Analysis: Analyze the distribution of product categories, sizes, and quantities sold to identify popular products.

```python
Qty = df1.groupby('Category')['Qty'].sum().tolist()
category = df1.Category.unique().tolist()
df2 = pd.DataFrame({"Category" : category, "Qty" : Qty})
df2.sort_values(by = 'Qty', ascending=False, inplace = True)
ax = sns.barplot(x = df2['Category'], y = df2['Qty'])
for i in ax.containers:
    ax.bar_label(i)
```

- In this bar chart, we can clearly see the overall sales of each category of product.
- The most bought product is `Socks`, while the least bought product is `Watch`.

## 6. Business Insights: Provide actionable insights and recommendations based on the analysis to optimize sales strategies, improve customer satisfaction, and enhance overall business performance.

```python
state = df1['ship-state'].unique().tolist()[1]
df_show = pd.DataFrame(df1[df1['ship-state'] == state].groupby('Size')['Category'].value_counts())
df_show
```

- As we can see here, In the state of Maharashtra, the most bought product is the Shirt of size 'XL'.
- Similarly we can find the popular product in all the states.
- To expand the business, we should increase the production of high-demand products and focus on selling them in the states where their sales are the highest.

```python
plt.figure(figsize = (15, 15))
states = df1['ship-state'].unique().tolist()
x1 = []
y = []
for state in states:
    df_l = pd.DataFrame(df1[df1['ship-state'] == state].groupby('Size')['Category'].value_counts())
    x = pd.DataFrame(df_l.iloc[np.argmax(df_l)])
    y.append(df_l.iloc[np.argmax(df_l)][0])
    x1.append(f'{state} : {x.columns.values[0][0]}, {x.columns.values[0][1]}')
    
ax = sns.barplot(y = x1, x = y)
for i in ax.containers:
    ax.bar_label(i)
```

- Hence Here is the list of products that are popular in each states which can be manufactured and sold more often in that state to generate more revenue.
- This is a business strategy followed by most of the firms
